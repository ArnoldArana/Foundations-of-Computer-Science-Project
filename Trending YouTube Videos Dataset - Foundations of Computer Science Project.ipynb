## Files with encoding issues

```python
#In this cell we are reading the datasets with encoding issues
JPvideos = pd.read_csv(r"C:\Users\arnol\OneDrive\Desktop\TrendingYT_dataset\trendingYT_csv_files\JPvideos.csv", encoding="latin-1")
KRvideos = pd.read_csv(r"C:\Users\arnol\OneDrive\Desktop\TrendingYT_dataset\trendingYT_csv_files\KRvideos.csv", encoding="latin-1")
MXvideos = pd.read_csv(r"C:\Users\arnol\OneDrive\Desktop\TrendingYT_dataset\trendingYT_csv_files\MXvideos.csv", encoding="latin-1")
RUvideos = pd.read_csv(r"C:\Users\arnol\OneDrive\Desktop\TrendingYT_dataset\trendingYT_csv_files\RUvideos.csv", encoding="latin-1")

#we applied the encoding latin-1 so that we can use a function developed by Chatgpt to fix the encoding problem
#----------------------------------------------------------------------------------------------------


# Function to fix "mojibake" (like â€™, Ã±, or Â£)
def fix_mojibake(corrupted_string):
    """Fixes strings that were UTF-8 but read as Latin-1."""
    if pd.isna(corrupted_string) or not isinstance(corrupted_string, str): 
        return corrupted_string #If this value is missing OR if it is anything other than a string, leave it alone and just return it as-is, it's like a guard..
    try:
        return corrupted_string.encode('latin-1').decode('utf-8')
        # Step 1: Convert the "wrongg" string back into raw bytes using Latin-1.
        # Step 2: Re-interpret those same bytes as UTF-8 to get the correct characters.
    except Exception:
        #If anything goes wrong during this conversion just stop trying and give us the original data back.
        return corrupted_string
        
#--------------------------------------------------------------------------------------------------------------------        

# Generic function to apply the fix to a DataFrame
def fix_encoding_dataframe(df, text_columns=['title', 'channel_title', 'tags', 'description']):
    """
    Applies the mojibake fix to the specified text columns in a DataFrame.
    """
    for col in text_columns:
        if col in df.columns:
            df[col] = df[col].apply(fix_mojibake)
        else:
            print(f"Warning: Column '{col}' not found in the DataFrame.")
    return df

# List of datasets to fix
datasets = [JPvideos, KRvideos, MXvideos, RUvideos]

# Apply the function to all DataFrames
datasets = [fix_encoding_dataframe(df) for df in datasets]

print("Mojibake correction completed on all datasets!")
```

## 5)Add a like_ratio column storing the ratio between the number of likes and of dislikes

```python
#firstly we made a function in order to do the ratio only if dislikes takes a value > 0. 
#If dislikes is equal to zero the function returns NA
#pd.NA is a pandas way to represent missing values, we cannot write only NA beacuse it does not exist in python
def ratio(row):
    if row['dislikes'] != 0:
        return row['likes']/row['dislikes']
    return pd.NA

#we add like_ratio column to concatdataframe dataframe
#arguments
# ratio = name of our function
# axis = 1 --> we want to apply our function to each row in order to to the ratio for each row
concatdataframe['like_ratio'] = concatdataframe.apply(ratio, axis = 1)
concatdataframe

# concatdataframe['like_ratio'] = concatdataframe['likes'] / concatdataframe['dislikes']
# Of course there is another way to solve this but we wanted to apply a function in order to
# solve the problem of dislikes = 0
```

## 6)Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
concatdataframe['publish_time']  #dtype = object
```

```python
import datetime #python library
#concatdataframe['publish_time'] series dtype object --> this is why we convert it in a time to be readable by pandas
concatdataframe['publish_time'] = pd.to_datetime(concatdataframe['publish_time'], format='mixed')
#to_datetime help us to transform object in datetimes in order to apply datetime functions
# format='mixed' means that the datetime values may have different formats
#exploiting "format = 'mixed' " argument we can transform every different format to a unique format

# dt.floor --> to round datetime object down to a specified frequency --> dt.floor(frequency)
# if we want to round to hour, let's suppose 12:20:03 we will get 12:00:00
# in our case we wanted to round to minutes to the nearest multiple of 10
concatdataframe['interval_start'] = concatdataframe['publish_time'].dt.floor('10min')  #pandas function which requires a series
# timedelta is very usefull to do operations with time, in our case we have exploited timedelta to ad 10 minutes to the initial interval
concatdataframe['interval_end'] = concatdataframe['interval_start'] + datetime.timedelta(minutes=10)  
concatdataframe
```

```python
#as we can see tags contains a string with several tags
#we have to separate every tag
#we start removing " and | by making a python function
def splitter_list(row):
    if row['tags'] != '[none]':
        return [t.strip() for t in row['tags'].replace('"','').split('|')]  #split separates the tags and create a list
    return ['[none]'] 

#we put none in a list in order to after apply the explode function 

video_idtags['tags'] = video_idtags.apply(splitter_list, axis = 1)
```
